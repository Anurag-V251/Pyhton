```python
np.random.seed(19680801) #seed the random number generator
data = {'a': np.arange(50),
        'c': np.random.randint(0,50,50),
        'd': np.random.randn(50)}
data['b'] = data['a']+10*np.random.randn(50)
data['d'] = np.abs(data['d'])*100

fig, ax = plt.subplots(figsize=(5,2.7))
ax.scatter('a','b',c='c',s='d',data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```

```python
x = np.linspace(0,2,100) # sample data

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,x,label='linear')
ax.plot(x,x**2,label='quadratic')
ax.plot(x,x**3,label='cubic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title("sample plot")
ax.legend()
```

```python
def my_plotter(ax,data1,data2,param_dict):
    out = ax.plot(data1,data2, **param_dict)
    return out

data1,data2,data3,data4 = np.random.randn(4,100)

fig,(ax1,ax2) = plt.subplots(1,2,figsize=(5,2.7))
my_plotter(ax1,data1,data2,{'marker':'x'})
my_plotter(ax2,data1,data2,{'marker':'o'})
```

```python
x = np.linspace(0,4*np.pi,1001)
fig,ax = plt.subplots(figsize=(5,2.7),dpi = 100)
ax.plot(x,np.sin(x),label='sine')
ax.plot(x,np.cos(x),label='cos',ls='--')
plt.xlim(0,2*np.pi)
ax.set_xticks(np.arange(0, 2*np.pi+0.01, np.pi/4))
labels = [r'$0$', r'$\pi/4$', r'$\pi/2$', r'$3\pi/4$', r'$\pi$',
          r'$5\pi/4$', r'$3\pi/2$', r'$7\pi/4$', r'$2\pi$']
ax.set_xticklabels(labels)
ax.legend()
ax.set_xlabel('X')
ax.set_ylabel('Function value')
plt.show()
```

```python
x = np.linspace(0,10,100)
fig,ax = plt.subplots(figsize=(5,2.7),dpi=150)
ax.plot(x,np.sin(x),label='sine')



def summation(x,end):
    sum = 0
    n = 1   
    while(n<=end):
            sum = sum + (((-1)**(n+1))*((x)**(2*n-1)))/np.math.factorial(2*n-1)
            n = n+1
    return sum
# summation(0,1)
# def sum(x,end):
#     i = 0
#     sum = 0
#     while(i<end):
#         sum = sum + x**(i)
#         i = i+1
#         print(sum)
#     return sum
# print(sum(1,3))
def my_function(end):
    data =[]
    for i in x:
        data.append(summation(i,end))
    return data
data1 = my_function(8)
ax.plot(x,data1,label='exp',ls='-.',color='red')
ax.set_xlim(left=0,right=2*np.pi)
ax.set_ylim(bottom=-1,top=1)
ax.legend()
```

```python
fig, ax = plt.subplots(figsize=(5,2.7))
x = np.linspace(0,5,100)
data = []
for i in x:
    data.append(np.math.exp(i))
ax.plot(x,data,label='exponential',color='red')
ax.legend(loc='best')

#now making e^x graph with its exapnsion

def expansion(x,end):
    sum = 0
    n = 1
    while(n<=end):
        sum= sum + (x**(n-1))/np.math.factorial(n-1)
        n = n+1
    return sum

def data_cal(x_values,end):
    data1 =[]
    for i in x_values:
        data1.append(expansion(i,end))
    return data1

data = data_cal(x,8)
ax.plot(x,data,label='expansion')
ax.legend(loc='best')

plt.show()
```
