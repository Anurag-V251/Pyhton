```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
k,t,y = 1,0.0,5.0
y0 = y
tt,yy=[],[]

tf =10

dt = 0.01

def f(t,y,k):
    return -k*y

while t<=tf:
    tt.append(t)
    yy.append(y)
    y = y+dt*f(t,y,k)
    t += dt
    
#calculation of exact result
tt = np.array(tt)
exact = y0*np.exp(-k*tt)


plt.subplot(2,1,1)
plt.plot(tt,yy,label="dt=%.4f"%(dt),ls='--')
plt.plot(tt,y0*np.exp(-k*tt),'k',label="exact solution")
plt.xlabel("time")
plt.ylabel("y")
plt.legend(loc='best')

plt.subplot(2,1,2)
diff = exact-yy
plt.plot(tt,diff,label="absolute error")
plt.xlabel("time")
plt.ylabel("error")
plt.legend(loc='best')
plt.title('Ode by euler method')
```
